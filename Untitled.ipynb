```python
%matplotlib inline
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font = ["Meiryo", "Yu Gothic", "Hiragino Maru Gothic Pro"])

#ゴルトンボード表示関数：段数球数を指定する
def galton(steps, count) :
    #球が落ちた位置を入れるからのリストを用意する
    ans = []
    #指定された球だけ繰り返す
    for i in range(count):
        #球を落とす最初の位置を50にする
        val = 50
        #指定された段数繰り返す
        for j in range(steps):
            #0か1のランダムで0なら-1、1なら＋1する
            if random.randint(0,1) == 0:
                val = val - 1
            else :
                val = val + 1
        #最終的に球が落ちた位置をリストに追加する
        ans.append(val)
        
    #落下した結果のリストをデータフレームにして
    df = pd.DataFrame(ans)
    #0列目(落とした結果の列)をヒストグラムで表示
    df[0].plot.hist()
    plt.title(str(steps) + "段：" + str(count) + "個")
    plt.ylabel("")
    plt.show()
galton(500,10000)
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm

sns.set(font = ["Meiryo", "Yu Gothic", "Hiragino Maru Gothic Pro"])

df = pd.DataFrame( {
    "A":np.random.randint(0,100,15),
    "B":np.random.normal(50,10,15)
})

sns.distplot(df["A"], fit = norm, fit_kws = {"color":"red"})
plt.title("偏りのないランダムな値")
plt.show()
```

```python
from scipy.stats import norm

scorelist = [60,70,80]

for score in scorelist:
    cdf = norm.cdf(x = score, loc = 50, scale = 10)
    print(score, (1-cdf)* 100)
```

```python
from scipy.stats import norm

perlist = [0.1586, 0.02275, 0.00134]

for per in perlist:
    ppf = norm.ppf( q = (1-per), loc = 50, scale = 10)
    print(per, ppf)
```
